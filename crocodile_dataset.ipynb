## Exploratory data analysis (eda)

### Subtask:
Visualize the distribution of key variables using histograms, box plots, etc.


**Reasoning**:
Create histograms and box plots for the numerical columns 'Observed Length (m)' and 'Observed Weight (kg)' to visualize their distributions and identify outliers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(15, 10))

# Histogram for Observed Length (m)
plt.subplot(2, 2, 1)
sns.histplot(df['Observed Length (m)'], kde=True)
plt.title('Distribution of Observed Length (m)')

# Box plot for Observed Length (m)
plt.subplot(2, 2, 2)
sns.boxplot(x=df['Observed Length (m)'])
plt.title('Box Plot of Observed Length (m)')

# Histogram for Observed Weight (kg)
plt.subplot(2, 2, 3)
sns.histplot(df['Observed Weight (kg)'], kde=True)
plt.title('Distribution of Observed Weight (kg)')

# Box plot for Observed Weight (kg)
plt.subplot(2, 2, 4)
sns.boxplot(x=df['Observed Weight (kg)'])
plt.title('Box Plot of Observed Weight (kg)')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Create count plots for the categorical variables 'Common Name', 'Conservation Status', and 'Habitat Type' to visualize their frequencies.



```python
# Set figure size
plt.figure(figsize=(18, 6))

# Count plot for Common Name (showing top 10)
plt.subplot(1, 3, 1)
sns.countplot(y=df['Common Name'], order=df['Common Name'].value_counts().index[:10], palette='viridis')
plt.title('Frequency of Top 10 Common Names')

# Count plot for Conservation Status
plt.subplot(1, 3, 2)
sns.countplot(y=df['Conservation Status'], order=df['Conservation Status'].value_counts().index, palette='viridis')
plt.title('Frequency of Conservation Status')

# Count plot for Habitat Type (showing top 10)
plt.subplot(1, 3, 3)
sns.countplot(y=df['Habitat Type'], order=df['Habitat Type'].value_counts().index[:10], palette='viridis')
plt.title('Frequency of Top 10 Habitat Types')

plt.tight_layout()
plt.show()
```

## Exploratory data analysis (eda)

### Subtask:
Explore relationships between variables using scatter plots, correlation matrices, etc.


**Reasoning**:
Calculate the correlation matrix for the numerical columns and create a heatmap visualization.



```python
# Calculate the correlation matrix
numerical_cols = ['Observed Length (m)', 'Observed Weight (kg)']
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Observed Length and Weight')
plt.show()
```

**Reasoning**:
Generate a scatter plot to visualize the relationship between `Observed Length (m)` and `Observed Weight (kg)`.



```python
# Generate a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Observed Length (m)', y='Observed Weight (kg)', data=df)
plt.title('Scatter Plot of Observed Length vs. Observed Weight')
plt.xlabel('Observed Length (m)')
plt.ylabel('Observed Weight (kg)')
plt.show()
```

## Data visualization

### Subtask:
Create relevant visualizations to support the findings from EDA, descriptive, and potentially prescriptive analysis.


**Reasoning**:
Create the requested visualizations to support the findings from the previous analysis steps. This includes bar plots for categorical distributions, box/violin plots for comparing numerical variables across categories, and a scatter plot for the relationship between length and weight.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar plot for Conservation Status distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='Conservation Status', data=df, order=df['Conservation Status'].value_counts().index, palette='viridis')
plt.title('Distribution of Conservation Status')
plt.xlabel('Count')
plt.ylabel('Conservation Status')
plt.show()

# 2. Bar plot for top 10 Habitat Type distribution
plt.figure(figsize=(10, 6))
top_10_habitats = df['Habitat Type'].value_counts().nlargest(10).index
sns.countplot(y='Habitat Type', data=df, order=top_10_habitats, palette='viridis')
plt.title('Top 10 Habitat Types')
plt.xlabel('Count')
plt.ylabel('Habitat Type')
plt.show()

# 3. Box plot comparing Observed Weight (kg) across different Conservation Status
plt.figure(figsize=(12, 8))
sns.boxplot(x='Observed Weight (kg)', y='Conservation Status', data=df, order=df['Conservation Status'].value_counts().index, palette='viridis')
plt.title('Observed Weight (kg) by Conservation Status')
plt.xlabel('Observed Weight (kg)')
plt.ylabel('Conservation Status')
plt.show()

# 4. Box plot comparing Observed Length (m) across different Conservation Status
plt.figure(figsize=(12, 8))
sns.boxplot(x='Observed Length (m)', y='Conservation Status', data=df, order=df['Conservation Status'].value_counts().index, palette='viridis')
plt.title('Observed Length (m) by Conservation Status')
plt.xlabel('Observed Length (m)')
plt.ylabel('Conservation Status')
plt.show()

# 5. Scatter plot of Observed Length (m) vs Observed Weight (kg) with Conservation Status as hue
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Observed Length (m)', y='Observed Weight (kg)', hue='Conservation Status', data=df, palette='viridis', alpha=0.7)
plt.title('Observed Length vs. Observed Weight by Conservation Status')
plt.xlabel('Observed Length (m)')
plt.ylabel('Observed Weight (kg)')
plt.legend(title='Conservation Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
