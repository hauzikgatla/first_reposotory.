## Load the data

### Subtask:
Load the data from "/content/Netflix Dataset.csv" into a pandas DataFrame.


**Reasoning**:
Load the data into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv('/content/Netflix Dataset.csv')
display(df.head())
```

## Perform exploratory data analysis (eda)

### Subtask:
Explore the dataset to understand its structure, identify missing values, and gain initial insights. This includes checking data types, summary statistics, and identifying potential issues.


**Reasoning**:
Perform exploratory data analysis steps: check data types, missing values, summary statistics, and unique values for initial insights into the dataset.



```python
df.info()
display(df.isnull().sum())
display(df.describe())
display(df.nunique())
```

## Perform descriptive analysis

### Subtask:
Calculate descriptive statistics for relevant columns to summarize the main features of the dataset.


**Reasoning**:
Calculate value counts for 'Category' and 'Rating' columns and identify the top 10 frequent values for 'Director', 'Cast', and 'Country' columns to summarize the data distribution for these features.



```python
category_counts = df['Category'].value_counts()
print("Value counts for 'Category':")
display(category_counts)

rating_counts = df['Rating'].value_counts()
print("\nValue counts for 'Rating':")
display(rating_counts)

top_directors = df['Director'].value_counts().nlargest(10)
print("\nTop 10 Directors:")
display(top_directors)

top_cast = df['Cast'].value_counts().nlargest(10)
print("\nTop 10 Casts:")
display(top_cast)

top_countries = df['Country'].value_counts().nlargest(10)
print("\nTop 10 Countries:")
display(top_countries)
```

## Perform prescriptive analysis

### Subtask:
Based on the insights gained from EDA and descriptive analysis, identify potential areas for prescriptive analysis. This could involve analyzing trends, identifying popular content types, or exploring relationships between features.


**Reasoning**:
I will analyze the 'Release_Date', 'Category', 'Rating', 'Duration', and 'Type' columns to identify trends, relationships, and popular content types for prescriptive analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Analyze 'Release_Date' trends
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Release_Year'] = df['Release_Date'].dt.year.astype('Int64') # Use Int64 to handle NaNs
release_year_counts = df['Release_Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=release_year_counts.index, y=release_year_counts.values)
plt.title('Content Release Trends Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()

# 2. Investigate relationship between 'Category' and 'Rating'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Rating', hue='Category', palette='viridis')
plt.title('Relationship between Category and Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()

# 3. Explore 'Duration' for movies and TV shows
# Separate movies and TV shows
movies_df = df[df['Category'] == 'Movie'].copy()
tv_shows_df = df[df['Category'] == 'TV Show'].copy()

# Extract duration for movies (assuming format is 'X min')
movies_df['Duration_minutes'] = movies_df['Duration'].str.replace(' min', '').astype(float)

# Extract number of seasons for TV shows (assuming format is 'X Season(s)')
tv_shows_df['Duration_seasons'] = tv_shows_df['Duration'].str.replace(' Seasons?', '', regex=True).astype(float)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(movies_df['Duration_minutes'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations (minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.histplot(tv_shows_df['Duration_seasons'].dropna(), bins=15, kde=False, color='lightcoral')
plt.title('Distribution of TV Show Durations (seasons)')
plt.xlabel('Duration (seasons)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()


# 4. Examine 'Type' to identify popular content types and their distribution
# The 'Type' column contains comma-separated values, need to split and count
types_list = df['Type'].dropna().str.split(', ').explode()
type_counts = types_list.value_counts().nlargest(15) # Top 15 types

plt.figure(figsize=(12, 7))
sns.barplot(x=type_counts.index, y=type_counts.values, palette='viridis')
plt.title('Top 15 Popular Content Types')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize the data

### Subtask:
Visualize the distribution of content by release year and the top 10 countries with the most content.


**Reasoning**:
Create bar plots to visualize the distribution of content by release year and the top 10 countries as requested by the subtask.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar plot for content distribution by release year
plt.figure(figsize=(12, 6))
sns.barplot(x=release_year_counts.index, y=release_year_counts.values, palette='viridis')
plt.title('Content Distribution by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Bar plot for top 10 countries with the most content
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 10 Countries with the Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Present the findings

Summarize the key findings from the analysis and visualizations.


**Reasoning**:
Summarize the key findings from the analysis and visualizations based on the previous steps.



```python
print("Key Findings from the Netflix Dataset Analysis:\n")

print("1. Content Categories:")
print(f"- The dataset contains significantly more Movies ({category_counts['Movie']}) than TV Shows ({category_counts['TV Show']}).\n")

print("2. Popular Content Ratings and Relationship with Categories:")
print(f"- The most frequent content ratings overall are:\n{rating_counts.nlargest(5)}\n")
print("- The analysis and visualizations showed that certain ratings are more prevalent for specific categories (e.g., TV-MA and TV-14 are common for both, while R and PG-13 are more frequent for Movies, and TV-Y and TV-Y7 are common for TV Shows).\n")

print("3. Trends in Content Release Over the Years:")
print("- There has been a clear upward trend in the number of content titles released over the years, with a significant increase in recent years (e.g., 2019 and 2020 had the highest number of releases).\n")

print("4. Distribution of Movie Durations and TV Show Seasons:")
print("- Movie durations tend to be concentrated between 80 and 120 minutes, with a peak around 90-100 minutes.")
print("- TV shows are predominantly one season long, with a rapid decrease in the number of shows with more seasons.\n")

print("5. Most Frequent Content Types:")
print(f"- The top content types are:\n{type_counts}\n")
print("- 'International Movies' and 'Dramas' are the most frequent content types, indicating a strong presence of international and dramatic content.\n")

print("6. Top Countries Contributing Content:")
print(f"- The top 10 countries with the most content are:\n{top_countries}\n")
print("- The United States is the leading country in terms of content contribution, followed by India and the United Kingdom.\n")
```
